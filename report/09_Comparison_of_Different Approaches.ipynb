# Appendix B: Comparison of Models


#### Individual Analysis of Each Approach

1. **Baseline Model (Approach 1)**
   - **BLEU Scores:**
    - The baseline model achieved high BLEU scores for most responses, with an average of **0.9529**, indicating that it closely matches the reference text in structure and content.
   - **Cosine Similarity Scores:**
    - The cosine similarity scores were moderate, averaging **0.6391**, which suggests decent semantic alignment with the reference responses.
   - **Summary:**
    - The baseline model performs well in terms of structural accuracy (high BLEU) but shows room for improvement in semantic similarity.

2. **Fine-Tuned Baseline Model (Approach 2)**
   - **BLEU Scores:**
    - The fine-tuned model had significantly lower BLEU scores, averaging **0.0867**.
    - The model struggled to produce responses with a structure closely matching the reference text.
   - **Cosine Similarity Scores:**
    - Cosine similarity scores also dropped, with an average of **0.2065**, indicating weak semantic alignment.
   - **Summary:**
    - Fine-tuning negatively impacted both BLEU and Cosine Similarity scores, reducing the structural and semantic accuracy of responses.

3. **CNN-Based Model (Approach 3)**
   - **BLEU Scores:**
    - The CNN-based model also showed low BLEU scores, averaging **0.0114**, suggesting limited structural similarity to the reference text.
   - **Cosine Similarity Scores:**
    - Cosine similarity scores, however, were consistently **1.0000** for all responses.
    - This suggests that while responses might not be structurally similar, they are semantically very close to the references.
   - **Summary:**
    - The CNN model achieves perfect semantic similarity but struggles with structural alignment, as indicated by the low BLEU score.

4. **MLP-Based Model (Approach 4)**
   - **BLEU Scores:**
    - The MLP-based model had the lowest BLEU score, averaging **0.0083**, indicating minimal structural similarity.
   - **Cosine Similarity Scores:**
    - Similar to the CNN model, the cosine similarity scores were consistently **1.0000**.
   - **Summary:**
    - The MLP model performs poorly in structural alignment (low BLEU) but achieves perfect semantic similarity (high cosine similarity).
